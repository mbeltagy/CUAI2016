```julia
fx=Uniform(-1,1)
```

```julia
nsamples=1_000_000
y=rand(fx,nsamples).^2;
```

```julia
ax=subplot(1,1,1)
ax[:hist](y,30);
```

This is fine, but we want to calculate probability. So we use a lower level function to do that

```julia
h=fit(Histogram,y,nbins=50)
```

We have to scale the weights (or counts to be exact) by $\frac{nsamples}{nbins}$. We verify that the equation is then correct. We worked that out to be $\frac{1}{2\sqrt{y}}$

```julia
y=0:0.01:1
py=1./(2*sqrt(y))
plot(h.edges[1][1:end-1]+0.01, h.weights*50/nsamples,"*",y,py)
```

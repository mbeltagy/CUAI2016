```julia
gm=Geometric(0.15)
```

```julia
function calcP(nreps)
    countD=0
    count=0
    for i=1:nreps
        n=rand(gm)
        if n>10 && n<16
            countD+=1
            if n==12
                count+=1
            end
        end
    end
    count/countD
end
            
```

```julia
calcP(10000000)
```

```julia
denom=sum(pdf(gm,i) for i=11:15)
```

```julia
numerator=pdf(gm,12)
```

```julia
numerator/denom
```
